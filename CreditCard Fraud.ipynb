```python
import h2o
```

```python
h2o.init(ip="localhost", port=54323)
```

```python
data = h2o.import_file('./creditcard.csv')
```

```python
data.head()
```

```python
data[data["Class"] == 1]
```

```python
data[data["Class"] == 0]
```

*Determine Predictors and Response Column*

```python
x = data.columns
y = "Class"# response Column

x.remove(y)
x.remove("Time")
```

*Splitting Frames*

```python
train,test,validation = data.split_frame(ratios=[0.7, 0.1])
```

*Setting the response column as factor for binary classification*

```python
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
```

*Run AutoML*

```python
from h2o.automl import H2OAutoML
aml = H2OAutoML(max_runtime_secs = 2*60)
aml.train(x = x, y = y,
          training_frame = train, leaderboard_frame = validation)
```

*Show Leaderboard Models*

```python
lb = aml.leaderboard
lb
```

*Select Leader Model*

```python
model = aml.leader
```

```python
model
```

```python
model.predict(test)
```

*perf show*

```python
perf_test = model.model_performance(test_data=test)
perf_test.plot()
```

```python
perf_test.confusion_matrix()
```

```python
preds=model.predict(test)
merged = test.cbind(preds)
merged_pred = merged[:, [ "Class","predict"]]
submission = merged_pred.as_data_frame()
submission.head()
```

```python
submission.to_csv('submission.csv', index=False)
```
